```python
import random
import math
import re
```

```python
def primos_ate_n(n):
    primos = []
    naturais_ate_n = [True] * n
    naturais_ate_n[0] = False
    naturais_ate_n[1] = False
    for i in range(2, n):
        for j in range(i * 2, n, i):
            naturais_ate_n[j] = False
    for i in range(len(naturais_ate_n)):
        if naturais_ate_n[i]:
            primos.append(i)
    return primos

def pega_primos(n):
    primos = primos_ate_n(n)
    p = 0
    q = 0
    while (p == q):
        id_i = random.randrange(0, len(primos) - 1)
        id_2 = random.randrange(0, len(primos) - 1)
        p = primos[id_i]
        q = primos[id_2]
    return {"p": p, "q": q}
            
```

```python
def gera_chaves(m):
    p = pega_primos(m)["p"]
    q = pega_primos(m)["q"]
 
    n = p * q
    fi = (p - 1) * (q - 1)
 
    chaves_pub = []
    
    for i in range(2,fi):
        if (math.gcd(i, fi) == 1):
            chaves_pub.append(i)
 
    chaves_priv = dict()
    for i in chaves_pub:
        di = 0
        while True:
            if ( (di * i) % fi == 1):
                chaves_priv[str(i)] = di
                break
            di += 1
 
    return {"n": n, "p": p, "q": q, "chaves_pub": chaves_pub, "chaves_priv": chaves_priv}

def criptografa(msg, n, e):
    msg_crip = []
    for carac in msg:
        msg_crip.append(str( (ord(carac))**(e) % n))
    return "_".join(msg_crip)

def descriptografa(msg_crip, n, d):
    msg_crip = re.split('_', msg_crip)
    msg_desc = []
    for carac in msg_crip:
        msg_desc.append( chr(int(carac)**d % n) )
    return "".join(msg_desc)
```

```python
m = 300

chaves = gera_chaves(m)
n = chaves["n"]
p = chaves["p"]
q = chaves["q"]
chaves_pub = chaves["chaves_pub"]
chaves_priv = chaves["chaves_priv"]
```
